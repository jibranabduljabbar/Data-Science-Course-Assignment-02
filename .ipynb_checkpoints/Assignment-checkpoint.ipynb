```python
import pandas as pd 
```

```python
data = pd.read_csv("Pokemon.csv")
```

```python
data
```

```python
speed_mean = data["Speed"].mean()
hp_mean = data["HP"].mean()
```

```python
def set_speed(val):
    if (val < speed_mean):
        return "Speed Low"
    elif (val == speed_mean):
        return "Speed Neutral"
    else:
        return "Speed High"
    
def set_hp(val):
    if (val < hp_mean):
        return "HP Low"
    elif (val == hp_mean):
        return "HP Neutral"
    else:
        return "HP High"        
```

```python
data["Speed_High_Low"] = data["Speed"].apply(set_speed)
data["HP_High_Low"] = data["HP"].apply(set_hp)
```
